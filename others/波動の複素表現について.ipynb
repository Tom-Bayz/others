# 波動の複素表現について

### 高校までの波動の表現

高校までは波動を時間$t$の関数として$ A\cos(\omega t)$みたいに三角関数を使って表現していた.

```python
import numpy as np 
from matplotlib import pyplot as plt

x = np.linspace(-8,8,100)
cos_wave = np.cos(x)

plt.plot(x,cos_wave,color="blue",label="cos wave")
plt.legend()
plt.show()
```

わかりやすい. 波打ってる.

### 複素表現の導入
でも大学に入ったら, $Ae^{i\omega t}$ <- こうなった. これはオイラーの公式で展開すると, 

$Ae^{i\omega t} = A\cos(\omega t) + iA\sin(\omega t)$

これが$ A\cos(\omega t)$と同じことを表すらしい. いやいや, $iA\sin(\omega t)$が余分にくっついてるやん

```python
import numpy as np 
from matplotlib import pyplot as plt

x = np.linspace(-8,8,100)
cos_wave = np.cos(x)
sin_wave = np.sin(x)

plt.plot(x,sin_wave,"--",color="blue",label="$i\ \sin(\omega t)$")
plt.plot(x,cos_wave,color="blue",label="$\cos(\omega t)$")
plt.legend()
plt.show()
```

ここで大前提として,「自然界では実部が観測されるとする」らしい. それは知らんかった.

### じゃあなんでこんな形で使われるの?
$Ae^{i\omega t}$と$\cos(\omega t)$が同じなることはわかった. ではなぜわざわざ複素数を用いた形が使われるようになったのか? それにはいくつか理由がある.

#### 微分が容易
指数関数は微分が簡単なのはご存知の通り. 何回微分しても同じ形が残る. 

$\frac{d}{dt}e^{\omega t}=\omega e^{\omega t}$

複雑な波動を考える際には, 何かと微分が簡単なほうが嬉しいらしい. <br>
